```python
train_image=os.listdir(image_dir)
```

```python
a = []
for image_name in train_image:
    if image_name not in list(train_df["ImageId"]):
        a.append(image_name)

b= [0]*len(a)

c= [None]*len(a)

new_df = pd.DataFrame({"ImageId":a, "ClassId":b, "EncodedPixels":c})

final_df = pd.concat([train_df,new_df])
```
